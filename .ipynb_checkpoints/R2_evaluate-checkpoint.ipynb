```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from keras.models import load_model
import tensorflow as tf
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from sklearn.metrics import r2_score
import matplotlib.gridspec as gridspec
import matplotlib.font_manager as font_manager
```

```python
# Calculate the R2 of test of train dataset
# where testt = True, this function will calculate the R2 of Testset
# And testt = False, this function will calculate the R2 of Testset
# Output of the function are the R2 value and result of the prediction
def R2(material, model, testt = True):
    if testt == True:
        X_dataset = pd.read_csv('test_train/%s_X_test.csv' %material)
        X_dataset = X_dataset.iloc[: , 1:]
        Y_dataset = pd.read_csv('test_train/%s_y_test.csv' %material)
        sett = 'Test set'
    if testt == False:
        X_dataset = pd.read_csv('test_train/%s_X_train.csv' %material)
        X_dataset = X_dataset.iloc[: , 1:]
        Y_dataset = pd.read_csv('test_train/%s_y_train.csv' %material)
        sett = 'Train set'
    
    model_address = 'model_%s/%s_relu_%s.h5' %(material, material, model)
    model = load_model(model_address)
    y_predict = model.predict(X_dataset)
    Y = pd.DataFrame(y_predict, columns=['y_pred'])
    Y['y_real'] = Y_dataset['OD']
    
    data_address = 'original/%s.csv' %material
    df = pd.read_csv(data_address)
    data = df['OD'].tolist()
    max_value = max(data)
    min_value = min(data)
    
    # Revers the nomalization to the real value for plotting
    Y['y_real'] = Y['y_real'] *(max_value - min_value)
    Y['y_pred'] = Y['y_pred'] *(max_value - min_value)
    
    Y = Y.sort_values(by=['y_real'])
    
    r2 = r2_score(Y['y_real'], Y['y_pred'])
    return r2, Y
```

```python
# calculate the R2 of all model and collect it
# By the condition R2 of test and train more than 0.8
# Output is list of the model name that follow by the condition
def findR2(name):
    address = 'model_%s/loss' %(name)
    stock = []    
    for i in range(4, 31):
        for j in range(4, 31):
            for k in range(4, 31):
                filename = '%s/%s_hist_relu_%d_%d_%d.csv' %(address, name, i, j, k)
                try:
                    df = pd.read_csv(filename)
                except:
                    break
                model = '%d_%d_%d' %(i, j, k)
                # test and train dataset of all model is the same dataset
                # But performance of each model is diverge
                R2_test, Y1 = R2(name, model, testt = True)
                R2_train, Y2 = R2(name, model, testt = False)
                if (R2_test >= 0.8) and (R2_train >= 0.8):
                    model = '%d_%d_%d' %(i, j, k)
                    stock.append(model)
    return stock

# List of the top 100 model which lowest percent error
def findPE_top100(name):
    address = 'error_table/%s_error.csv' %(name)
    stock = []    
    df = pd.read_csv(address)
    df = df.head(200)
    neuron = df['neuron']
    Perror = df['Perror']
    for i in range(0,len(Perror)):
        node = neuron[i]
        node = node.replace('relu_', '')
        stock.append([Perror[i], node])
        stock.sort()
    return stock
```

```python
# The function for plotting the R2 of test and train data and plot of the training loss and validation loss
def plot_R2(r2test, r2train, Ytest, Ytrain, paper, model):
    
    font = font_manager.FontProperties(family='serif',style='normal', size=16)
    y_test = np.array(Ytest['y_real'])
    y_predd = np.array(Ytest['y_pred'])
    m, b = np.polyfit(y_test, y_predd, 1)
    regress = m*y_test + b
    
    gs = gridspec.GridSpec(2, 2,wspace = 0.1)
    fig = plt.figure(figsize=(16,16))
    
    ax1 = fig.add_subplot(gs[0, 0])
    ax1.scatter(y_test, y_predd, label = 'data')
    ax1.plot(y_test, regress, label = 'Regression line')
    ax1.plot(y_test, y_test, linestyle='dotted', label = 'Y = X')
    ax1.set_aspect(1)
    ax1.set_title('Test set : R\u00B2 = %0.5f' %(r2test), fontsize = 30, fontfamily = 'serif')
    ax1.set_xlabel('Actual diameter (nm)', fontsize = 16, fontfamily = 'serif') 
    ax1.set_ylabel('Predicted dimeter (nm)', fontsize = 16, fontfamily = 'serif')
    ax1.grid()
    ax1.legend(prop=font, loc = 4)
    ax1.text(125, 400, '(A)', size=30, ha='center', va='center',bbox ={'facecolor':'white', 'pad':10})
    
    
    y_test = np.array(Ytrain['y_real'])
    y_predd = np.array(Ytrain['y_pred'])
    m, b = np.polyfit(y_test, y_predd, 1)
    regress = m*y_test + b
    ax2 = fig.add_subplot(gs[0, 1])
    ax2.scatter(y_test, y_predd, label = 'data')
    ax2.plot(y_test, regress, label = 'Regression line')
    ax2.plot(y_test, y_test, linestyle='dotted', label = 'Y = X')
    ax2.set_aspect(1)
    ax2.set_title('Train set : R\u00B2 = %0.5f' %(r2train), fontsize = 30, fontfamily = 'serif')
    ax2.set_xlabel('Actual diameter (nm)', fontsize = 16, fontfamily = 'serif') 
    ax2.set_ylabel('Predicted dimeter (nm)', fontsize = 16, fontfamily = 'serif')
    ax2.grid()
    ax2.legend(prop=font, loc = 4)
    ax2.text(150, 500, '(B)', size=30, ha='center', va='center',bbox ={'facecolor':'white', 'pad':10})
    
    ax3 = fig.add_subplot(gs[1, :]) 
    address = 'model_%s/loss' %(paper)
    df = pd.read_csv('%s/%s_hist_relu_%s.csv' %(address,paper,model))
    loss = df['loss']
    val_loss = df['val_loss']                 
    epoch = np.arange(0,500)
    ax3.plot(epoch, loss, label='Train loss')
    ax3.plot(epoch, val_loss, label='Validation loss')
    ax3.set_ylim(0,0.1)
    ax3.set_title('Epoch vs Training and Validation loss',fontsize = 30, fontfamily = 'serif')
    ax3.set_xlabel('Epoch', fontsize = 22, fontfamily = 'serif') 
    ax3.set_ylabel('Loss (Mean square error)', fontsize = 22, fontfamily = 'serif')
    ax3.grid()    
    ax3.text(0, 0.09, '(C)', size=30, ha='center', va='center',bbox ={'facecolor':'white', 'pad':10})
    font = font_manager.FontProperties(family='serif',style='normal', size=24)
    ax3.legend(prop=font)                               
    
    fig.subplots_adjust(wspace = 0)
    ax1.tick_params(axis='both', which='major', labelsize=16)
    ax2.tick_params(axis='both', which='major', labelsize=16)
    ax3.tick_params(axis='both', which='major', labelsize=20)
    plt.show()
```

```python
# Example of the programe
name = 'Mixing'
model = '13_30_21'
r2test, Y1 = R2(name, model, testt = True)
r2train, Y2 = R2(name, model, testt = False)
plot_R2(r2test, r2train, Y1, Y2,name,model)
```
