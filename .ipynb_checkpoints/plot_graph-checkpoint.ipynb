```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from keras.models import load_model
import tensorflow as tf
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from sklearn.metrics import r2_score
```

```python
class electrospinning:
        
    def __init__(self, paper_name, validation, neuron):
        self.paper_name = paper_name
        self.validation = validation
        self.neuron = neuron
        self.parameters = ['norm_con', 'norm_vol', 'norm_dis', 'norm_feed']
    
    def file_name(self):
        if self.validation == True:
            address = 'validation/last_%s' %(self.paper_name)
        if self.validation == False:
            #if self.paper_name == 'norm_mix':
            address = 'last_last_test_%s' %(self.paper_name)    
            #else:
            #    address = 'nonvalidation/last_test_%s' %(self.paper_name)
            
        
        file = []
        for i in self.neuron:
            file.append('%s/%s_relu_%s.h5' %(address, self.paper_name, i))
        return file
    
    def data_linspace(self):
        csv = pd.read_csv('csv_file/%s.csv' %self.paper_name)
        value = []
        for i in self.parameters:
            x = csv[i]
            table = np.linspace(x.min(), x.max(), num = 100, endpoint = False)
            table = table.tolist()
            value.append(table)
        df = pd.DataFrame(value)
        df = df.transpose()
        df.columns = self.parameters
        return df
    
    def prediction_table(self):
        df = electrospinning.data_linspace(self)
        csv = pd.read_csv('csv_file/%s.csv' %self.paper_name)
        
        high = []
        mid = []
        low = []
        
        num = len(csv['norm_con'])
        lownum = 1
        highnum = num - 1
        midnum = num//2
        for i in self.parameters:
            data = csv[i]
            data = data.sort_values()
            data = data.tolist()
            high.append(data[highnum])
            mid.append(data[midnum])
            low.append(data[lownum])
        r = [high, mid, low]
        value = pd.DataFrame(r)
        value.columns = self.parameters
        
        ##### *50 #####
        
        tablee = [[], [], [], []]
        i = 0
        for j in self.parameters:
            
            for k in range(0,3):
                t = [value[j][k]]*100
                
                tablee[i].append(t)
            i += 1    
        return tablee
    
    def make_2Dtable(self, x, p1='mid', p2='mid', p3='mid'):
        check = []
        Parameter = [p1, p2, p3]
        for i in Parameter:
            if i == 'high':
                check.append(0)
            elif i == 'mid':
                check.append(1)
            else:
                check.append(2)
        
        ##### making table #####
        
        llinspace = electrospinning.data_linspace(self)
        constant = electrospinning.prediction_table(self)
        
        xx = llinspace[x]
        xx = xx.tolist()
        plot_table = []
        
        count = 0
        for i in range(0, 4):
            if x == self.parameters[i]:
                plot_table.append(xx)
            else:
                plot_table.append(constant[i][check[count]])
                count += 1
        
        df = pd.DataFrame(plot_table)
        df = df.transpose()
        df.columns = self.parameters
        
        return df
    
    def make_3Dtable(self, x, y, p1='mid', p2='mid'):
        check = []
        Parameter = [p1, p2]
        for i in Parameter:
            if i == 'high':
                check.append(0)
            elif i == 'mid':
                check.append(1)
            else:
                check.append(2)
        
        ##### making table #####
        
        llinspace = electrospinning.data_linspace(self)
        constant = electrospinning.prediction_table(self)
        
        xx, yy = llinspace[x], llinspace[y]
        xx, yy = xx.tolist(), yy.tolist()
        plot_table = []
        
        count = 0
        for i in range(0, 4):
            if x == self.parameters[i]:
                plot_table.append(xx*100)
            elif y == self.parameters[i]:
                plot_table.append(yy*100)
            else:
                plot_table.append(constant[i][check[count]]*100)
                count += 1
        
        df = pd.DataFrame(plot_table)
        df = df.transpose()
        df.columns = self.parameters
        
        y_table = []
        for i in yy:
            y_table += [i]*100
        
        df[y] = y_table
        #df.to_csv('test.csv')
        return df
    
    def delete_norm(self, x):
        count = 0
        strr = ''
        for i in x:
            if count >= 5:
                strr += i
            count += 1
        return strr
    
    def axis_label(self, axis):
        if axis == 'norm_con':
            return 'Concentration'
        elif axis == 'norm_vol':
            return 'Applied Voltage (kV)'
        elif axis == 'norm_dis':
            return 'Tip-to-Nozzle Distance (cm)'
        elif axis == 'norm_feed':
            return 'Feed rate (ml/hr)'
    
    def check (self, x):
        if x == 'norm_esnaashari':
            y = 1
        elif x == 'norm_faridi':
            y = 2
        elif x == 'norm_rabbi':
            y = 3
        elif x == 'norm_sadan':
            y = 4
        return y
    
    
    
    def plot_2Dgraph(self, x, p1='mid', p2='mid', p3='mid', savepic = True, savetable = True, mix = False):
        name = electrospinning.file_name(self)
        X_data = electrospinning.make_2Dtable(self, x, p1, p2, p3)
        xx = electrospinning.data_linspace(self)
        diameter = []
        
        paper = electrospinning.delete_norm(self, self.paper_name)
        
        unnorm = pd.read_csv('electrospinning/%s.csv' %paper)
        
        unnorm_x = electrospinning.delete_norm(self, x)
        
        real_max_value = unnorm[unnorm_x].max()
        real_min_value = unnorm[unnorm_x].min()
        delta = real_max_value - real_min_value
        x_for_plot = xx[x]*delta
        
        OD_max = unnorm['OD'].max()
        OD_min = unnorm['OD'].min()
        OD_delta = OD_max - OD_min
        ver = 0
        if mix != False:            
                
            mix_table = pd.read_csv('electrospinning/mix.csv')
            listt = list(mix_table.columns)
            listt.remove('paper')
            listt.remove('OD')
            
            for i in listt:
                name = 'norm_' + i
                mix_max = mix_table[i].max()
                mix_min = mix_table[i].min()
                mix_delta = mix_max - mix_min
                
                x_max = unnorm[i].max()
                x_min = unnorm[i].min()
                x_delta = x_max - x_min
                
                X_data[name] = X_data[name]*x_delta/mix_delta
            
            real_max_value = unnorm[unnorm_x].max()
            real_min_value = unnorm[unnorm_x].min()
            delta = real_max_value - real_min_value
            
            
            x_for_plot = xx[x]*delta
            
            OD_max = mix_table['OD'].max()
            OD_min = mix_table['OD'].min()
            OD_delta = OD_max - OD_min
            if mix == 'label':
                
                order = electrospinning.check(self, self.paper_name)
                order = [order] * 100
                order = pd.DataFrame(order)
                order.columns = ['paper']
                X_data['paper'] = order['paper']
                titles = list(X_data.columns)
                titles[0], titles[1], titles[2], titles[3], titles[4] = titles[4], titles[0], titles[1], titles[2], titles[3] 
                X_data = X_data[titles]
                
                mixx = electrospinning('norm_mix', self.validation, self.neuron)
                name = mixx.file_name()
                ver = 1
            if mix == 'nonlabel':
                mixx = electrospinning('norm_mix_nonlabel', self.validation, self.neuron)
                name = mixx.file_name()
                ver = 2
        
        result = x_for_plot.tolist()
        result = pd.DataFrame(result, columns = ['X'])                                
        
        
        for i in range(0, len(self.neuron)):
            model = load_model(name[i])
            y_predict = model.predict(X_data)
            y_plot = y_predict*OD_delta
            diameter.append(y_plot)
            labell = '%s_%s'%(self.paper_name, self.neuron[i])
            plt.plot(x_for_plot, y_plot, label = '%s' %labell)
            result['y_%d'%(i + 1)] = y_predict
        plt.xlabel('%s' %electrospinning.axis_label(self, x))
        plt.ylabel('Nanafiber diameters (nm)')
        plt.ylim(200, 520)
        plt.grid(which='both')
        #####
        plt.title('%s_%s'%(self.paper_name, x))
        plt.legend()
        if savepic == True:
            if mix != False:
                plt.savefig('save_pic/mix_%s_%s_%d%s%s%s.png'%(self.paper_name, x, 
                                                               ver,p1, p2, p3), transparent=True)
            else:
                plt.savefig('save_pic/%s_%s_%d%s%s%s.png'%(self.paper_name, x, 
                                                           ver,p1, p2, p3), transparent=True)
        plt.show()
        result['y_plot'] = y_plot
        if savetable == True:
            if mix != False:
                result.to_csv('save_table/mix_%s_%s_%d%s%s%s.csv'%(self.paper_name, x, ver,
                                                                   p1, p2, p3), index = False)
            else:
                result.to_csv('save_table/mix_%s_%s_%d%s%s%s.csv'%(self.paper_name, x, ver,
                                                                   p1, p2, p3), index = False)
        #return X_data
        return result
    
    def plot_3Dsurface(self, x, y, p1='mid', p2='mid', mix = 'label', order=0, savepic = True, savetable = True):
        from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

        import matplotlib.pyplot as plt
        from matplotlib import cm
        from matplotlib.ticker import LinearLocator, FormatStrFormatter
        import numpy as np
        
        name = electrospinning.file_name(self)
        base_data = electrospinning.make_3Dtable(self, x, y, p1, p2)
        
        paper = electrospinning.delete_norm(self, self.paper_name)
        
        unnorm = pd.read_csv('electrospinning/%s.csv' %paper)
        unnorm_x = electrospinning.delete_norm(self, x)
        unnorm_y = electrospinning.delete_norm(self, y)
        
        x_real_max_value = unnorm[unnorm_x].max()
        x_real_min_value = unnorm[unnorm_x].min()
        x_delta = x_real_max_value - x_real_min_value
        x_for_plot = base_data[x]*x_delta
        
        y_real_max_value = unnorm[unnorm_y].max()
        y_real_min_value = unnorm[unnorm_y].min()
        y_delta = y_real_max_value - y_real_min_value
        y_for_plot = base_data[y]*y_delta
        
        OD_max = unnorm['OD'].max()
        OD_min = unnorm['OD'].min()
        OD_delta = OD_max - OD_min
        
        if mix != False:
            #model_address = 'last_last_test_norm_mix/norm_mix_relu_%s.csv' %self.neuron[order]
            mixxx = pd.read_csv('electrospinning/mix.csv')
            unnorm = pd.read_csv('electrospinning/%s.csv' %paper)
            for i in ['con', 'vol', 'dis', 'feed']:
                mixmax = mixxx[i].max()
                mixmin = mixxx[i].min()
                maxx = unnorm[i].max()
                minn = unnorm[i].min()
                base_data['norm_%s' %i] = base_data['norm_%s' %i]*(maxx - minn)/(mixmax - mixmin)
            
            #OD_max = mixxx['OD'].max()
            #OD_min = mixxx['OD'].min()
            #OD_delta = (OD_max - OD_min)
        
        
        
        model = load_model(name[order])
        z_predict = model.predict(base_data)
        z_plot = z_predict*OD_delta
        base_data['diameter'] = z_plot
        
        x1 = x_for_plot
        y1 = y_for_plot
        z1 = base_data['diameter'].tolist()
        print(len(x1), len(y1), len(z1))
        
        
        fig1 = plt.figure()
        ax1 = plt.axes(projection='3d')
        surf1 = ax1.plot_trisurf(x1, y1, z1, cmap='viridis', edgecolor='none');
        #ax1.view_init(79, 47)
        ax1.set_xlabel('%s' %electrospinning.axis_label(self, x))
        ax1.set_ylabel('%s' %electrospinning.axis_label(self, y))
        
        
        ax1.set_zlabel('Diameter(nm)')
        #fig1.colorbar(surf1, shrink=1, aspect=10)
        fig1.colorbar(surf1)
        plt.title('%s_%s_%s%s' %(unnorm_x, unnorm_y,p1,p2))
        if savepic == True :
            plt.savefig('save_pic/%s_%s_%s%s.png' %(unnorm_x, unnorm_y,p1,p2), transparent=True)
        plt.show()
        base_data[unnorm_x] = x_for_plot
        base_data[unnorm_y] = y_for_plot
        if savetable == True:
            base_data.to_csv('save_table/%s_%s_%s_%s%s.csv' %(self.paper_name,unnorm_x, unnorm_y,p1,p2), index = False)
        return base_data
```
